```python
import nltk
from nltk import sent_tokenize, word_tokenize
from string import punctuation
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer 
ps=PorterStemmer()
from nltk.stem import WordNetLemmatizer
wnl=WordNetLemmatizer()
```

```python
with open('brown_corpus_ca10.txt','r') as myfile:
  data=myfile.read().replace('\n','')
data2=data.replace("/","")
```

```python
for i,line in enumerate(data2.split('\n')):
  if i>10:
    break
  print(str(i)+" \t "+line)
```

```python
senttokenize=sent_tokenize(data2)
words_tokenized=word_tokenize(data2)
#print("Sentence Tokenized: ")
#for sent in senttokenize:
#  print(sent)
stopwords_en=set(stopwords.words('english'))
stopwords_en_withpunct=stopwords_en.union(set(punctuation))
single_tokenized_lower=list(map(str.lower,words_tokenized))
filtered_words=[]
for word in single_tokenized_lower:
  if word not in stopwords_en_withpunct:
    filtered_words.append(word)
#print("Stopwords Removed",filtered_words)
stemmed_words=[]
for word in filtered_words:
  stemmed_word="Stemmed word: "+ps.stem(word)
  lmmatized_word="Lemmatized word"+wnl.lemmatize(word)
  stemmed_words.append((stemmed_word,lmmatized_word))
print(stemmed_words)
```

```python
sentences=sent_tokenize(data2)
tokenize_sentence=[nltk.word_tokenize(sent) for sent in sentences]
tagged_sentence=[nltk.pos_tag(sentence) for sentence in tokenize_sentence]
print(tagged_sentence)
```

```python
chunked_sentence=nltk.ne_chunk_sents(tagged_sentence,binary=True)
def extractEntityNames(t):
  entity_names=[]
  if hasattr(t,'label') and t.label:
    if t.label() =='NE':
      entity_names.append(' '.join(child[0] for child in t))
    else:
      for child in t:
        entity_names.extend(extractEntityNames(child))
  return entity_names;

entity_names=[]
for tree in chunked_sentence:
  entity_names.extend(extractEntityNames(tree))
print(set(entity_names))
```
